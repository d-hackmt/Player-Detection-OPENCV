```python
import cv2
import numpy as np

# Reading the video
vidcap = cv2.VideoCapture('cutvideo.mp4')  # Replace 'path_to_video_file' with the actual path
success, image = vidcap.read()
count = 0
idx = 0

# Read the video frame by frame
while success:
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define color ranges
    lower_green = np.array([40, 40, 40])
    upper_green = np.array([70, 255, 255])
    lower_blue = np.array([110, 50, 50])
    upper_blue = np.array([130, 255, 255])
    lower_red = np.array([0, 31, 255])
    upper_red = np.array([176, 255, 255])
    lower_white = np.array([0, 0, 0])
    upper_white = np.array([0, 0, 255])

    # Create masks
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
    mask_red = cv2.inRange(hsv, lower_red, upper_red)
    mask_white = cv2.inRange(hsv, lower_white, upper_white)

    # Bitwise AND operation
    res_green = cv2.bitwise_and(image, image, mask=mask_green)
    res_blue = cv2.bitwise_and(image, image, mask=mask_blue)
    res_red = cv2.bitwise_and(image, image, mask=mask_red)
    res_white = cv2.bitwise_and(image, image, mask=mask_white)

    res_gray = cv2.cvtColor(res_green, cv2.COLOR_BGR2GRAY)
    kernel = np.ones((13, 13), np.uint8)
    thresh = cv2.threshold(res_gray, 127, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    font = cv2.FONT_HERSHEY_SIMPLEX

    for c in contours:
        x, y, w, h = cv2.boundingRect(c)
        if h >= 1.5 * w:
            if w > 15 and h >= 15:
                idx = idx + 1
                player_img = image[y:y + h, x:x + w]
                player_hsv = cv2.cvtColor(player_img, cv2.COLOR_BGR2HSV)
                nzCount_blue = cv2.countNonZero(cv2.inRange(player_hsv, lower_blue, upper_blue))
                nzCount_red = cv2.countNonZero(cv2.inRange(player_hsv, lower_red, upper_red))
                if nzCount_blue >= 20:
                    cv2.putText(image, 'France', (x - 2, y - 2), font, 0.8, (255, 0, 0), 2, cv2.LINE_AA)
                    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 3)
                if nzCount_red >= 20:
                    cv2.putText(image, 'Belgium', (x - 2, y - 2), font, 0.8, (0, 0, 255), 2, cv2.LINE_AA)
                    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 3)
        if 1 <= h <= 30 and 1 <= w <= 30:
            nzCount_white = cv2.countNonZero(cv2.inRange(cv2.cvtColor(image[y:y + h, x:x + w], cv2.COLOR_BGR2HSV),
                                                          lower_white, upper_white))
            if nzCount_white >= 3:
                cv2.putText(image, 'football', (x - 2, y - 2), font, 0.8, (0, 255, 0), 2, cv2.LINE_AA)
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)

    cv2.imwrite("./Cropped/frame%d.jpg" % count, res_green)
    print('Read a new frame:', success)  # save frame as JPEG file
    count += 1
    cv2.imshow('Match Detection', image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    success, image = vidcap.read()

vidcap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Reading the video
vidcap = cv2.VideoCapture('cutvideo.mp4')  
success, image = vidcap.read()
count = 0
idx = 0

# Resize factor
resize_factor = 0.48

# Read the video frame by frame
while success:
    # Resize frames
    resized_frame = cv2.resize(image, None, fx=resize_factor, fy=resize_factor)

    # Display original frame
    cv2.imshow('Original Frame', resized_frame)

    # Convert BGR to HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    resized_hsv = cv2.resize(hsv, None, fx=resize_factor, fy=resize_factor)
    cv2.imshow('HSV Image', resized_hsv)

    # Define color ranges
    lower_green = np.array([40, 40, 40])
    upper_green = np.array([70, 255, 255])
    lower_blue = np.array([110, 50, 50])
    upper_blue = np.array([130, 255, 255])
    lower_red = np.array([0, 31, 255])
    upper_red = np.array([176, 255, 255])
    lower_white = np.array([0, 0, 0])
    upper_white = np.array([0, 0, 255])

    # Create masks
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
    mask_red = cv2.inRange(hsv, lower_red, upper_red)
    mask_white = cv2.inRange(hsv, lower_white, upper_white)

    # Bitwise AND operation
    res_green = cv2.bitwise_and(image, image, mask=mask_green)
    res_blue = cv2.bitwise_and(image, image, mask=mask_blue)
    res_red = cv2.bitwise_and(image, image, mask=mask_red)
    res_white = cv2.bitwise_and(image, image, mask=mask_white)

    resized_res_green = cv2.resize(res_green, None, fx=resize_factor, fy=resize_factor)
    resized_res_blue = cv2.resize(res_blue, None, fx=resize_factor, fy=resize_factor)
    resized_res_red = cv2.resize(res_red, None, fx=resize_factor, fy=resize_factor)
    resized_res_white = cv2.resize(res_white, None, fx=resize_factor, fy=resize_factor)

    cv2.imshow('Green Mask', resized_res_green)
    cv2.imshow('Blue Mask', resized_res_blue)
    cv2.imshow('Red Mask', resized_res_red)
    cv2.imshow('White Mask', resized_res_white)

    res_gray = cv2.cvtColor(res_green, cv2.COLOR_BGR2GRAY)
    kernel = np.ones((13, 13), np.uint8)
    thresh = cv2.threshold(res_gray, 127, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    # Find contours
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    font = cv2.FONT_HERSHEY_SIMPLEX

    for c in contours:
        x, y, w, h = cv2.boundingRect(c)
        if h >= 1.5 * w:
            if w > 15 and h >= 15:
                idx = idx + 1
                player_img = image[y:y + h, x:x + w]
                player_hsv = cv2.cvtColor(player_img, cv2.COLOR_BGR2HSV)
                nzCount_blue = cv2.countNonZero(cv2.inRange(player_hsv, lower_blue, upper_blue))
                nzCount_red = cv2.countNonZero(cv2.inRange(player_hsv, lower_red, upper_red))
                if nzCount_blue >= 20:
                    cv2.putText(image, 'France', (x - 2, y - 2), font, 0.8, (255, 0, 0), 2, cv2.LINE_AA)
                    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 3)
                if nzCount_red >= 20:
                    cv2.putText(image, 'Belgium', (x - 2, y - 2), font, 0.8, (0, 0, 255), 2, cv2.LINE_AA)
                    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 3)
        if 1 <= h <= 30 and 1 <= w <= 30:
            nzCount_white = cv2.countNonZero(cv2.inRange(cv2.cvtColor(image[y:y + h, x:x + w], cv2.COLOR_BGR2HSV),
                                                          lower_white, upper_white))
            if nzCount_white >= 3:
                cv2.putText(image, 'football', (x - 2, y - 2), font, 0.8, (0, 255, 0), 2, cv2.LINE_AA)
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)

    resized_image = cv2.resize(image, None, fx=resize_factor, fy=resize_factor)
    resized_thresh = cv2.resize(thresh, None, fx=resize_factor, fy=resize_factor)

    cv2.imshow('Processed Frame', resized_image)
    cv2.imshow('Thresholded Image', resized_thresh)

    cv2.imwrite("./Cropped/frame%d.jpg" % count, res_green)
    print('Read a new frame:', success)  # save frame as JPEG file
    count += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    success, image = vidcap.read()

vidcap.release()
cv2.destroyAllWindows()
```
